# Introduction


### Cost function
Also known as objective function or loss function.The goal of training a model is to make predictions as accurate as possible.The cost function measures the error between predicted and actual values. The larger the error the higher the cost function. Example is MSE for regression tasks and cross entropy for classification tasks

### Gradient descent
Gradient descent is an optimization algorithm used to minimize the cost function in machine learning and deep learning models. It iteratively adjusts the model parameters to find the minimum value of the cost function, which represents the best fit for the training data. 

## Steps of Gradient Descent
- Initialize Parameters: Start with initial values for the model parameters, often set randomly.
- Compute the Cost: Calculate the cost (error) using the current parameters.
- Compute the Gradient: Calculate the gradient of the cost function with respect to each parameter. This involves partial derivatives that show the direction and rate of change of the cost with respect to each parameter.
- Update Parameters: Adjust the parameters by moving them in the direction opposite to the gradient. The size of the step is determined by the learning rate.
- Repeat: Repeat the steps of computing the cost, gradient, and updating parameters until the cost converges to a minimum or reaches a predefined number of iterations.

## Analogy: Finding the Lowest Point in a Valley
- Imagine you're standing on the side of a mountain in thick fog, and you want to find the lowest point (valley):

- Current Position: Your current position represents the current values of your model parameters.
- Foggy Landscape: The fog represents the complexity of the cost function landscape.
- Step Size: The size of your steps represents the learning rate.
- Slope: The steepness of the slope in any direction represents the gradient.
- Look Around: You look around to find the steepest direction downhill (compute the gradient).
- Take a Step: You take a step in that direction (update the parameters).
- Repeat: You repeat this process, taking one step at a time, moving downhill, until you reach the bottom of the valley (minimum cost).
